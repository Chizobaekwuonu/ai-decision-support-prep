```python
from transformers import pipeline
```

```python
agent = pipeline("text2text-generation", model="google/flan-t5-base")
```

```python
def decision_agent(question):
    prompt = (
        "You are an expert decision-support AI for global health supply chains. "
        "Always reply with a structured, multi-step answer.\n\n"
        "Your answer MUST include:\n"
        "1. An assessment of the situation\n"
        "2. Immediate actions to take\n"
        "3. Medium-term actions\n"
        "4. Risks and mitigation steps\n"
        "5. Any data required for verification\n\n"
        f"Question: {question}\n\n"
        "Provide a clear, detailed answer below:\n"
    )

    response = agent(prompt, max_length=350, do_sample=False)
    return response[0]["generated_text"]

```

```python
question1 = "There is a sudden spike in malaria cases in a rural region but drug stock is low. What should the supply chain team do?"
print(decision_agent(question1))
```

```python
question2 = "Donor funding is uncertain. How should a health supply chain plan stock levels for essential medicines?"
print(decision_agent(question2))
```

```python
question3 = "What are the key risks if vaccine deliveries are delayed by two weeks, and how can they be mitigated?"
print(decision_agent(question3))
```

I built a small natural-language decision-support agent using the FLAN-T5 Base model. It accepts a question about global health supply chains and returns structured recommendations. Building this helped me understand both the potential and limitations of LLM-based decision-making. The model can suggest sensible actions, but it has no real-time data or contextual understanding. This reinforces why the project’s focus on combining LLMs with forecasting, optimisation, and human-in-the-loop oversight is essential for safe and effective decision-support in health systems.
